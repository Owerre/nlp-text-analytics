# Feature Engineering
In this notebook, we will engineer count based and indicator features from the existing features. 

```python
%matplotlib inline

# ignore deprecated warning
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
```

 #  Load  data

```python
df = pd.read_csv("../data/data_processed.csv", sep = '\t')
print("Data size:", df.shape)
df.head()
```

# Day, month, and year count

```python
# Day count
df_day = pd.DataFrame(df.day_of_week.value_counts())
df_day.reset_index(level=0,inplace=True)
df_day.rename(columns = {'day_of_week':'wday_ct',  'index':'day_of_week'},inplace=True)

# Month count
df_mnth = pd.DataFrame(df.month.value_counts())
df_mnth.reset_index(level=0,inplace=True)
df_mnth.rename(columns = {'month':'month_ct',  'index':'month'},inplace=True)

# Year count
df_year = pd.DataFrame(df.year.value_counts())
df_year.reset_index(level=0,inplace=True)
df_year.rename(columns = {'year':'year_ct',  'index':'year'},inplace=True)
```

```python
# left join to the main dataframe
df = df.merge(df_day, how = 'left', on = 'day_of_week') 
df = df.merge(df_mnth, how = 'left', on = 'month')
df = df.merge(df_year, how = 'left', on = 'year') 
```

# Save as csv

```python
df.to_csv('../data/data_processed.csv', sep = '\t', index = False)
```
